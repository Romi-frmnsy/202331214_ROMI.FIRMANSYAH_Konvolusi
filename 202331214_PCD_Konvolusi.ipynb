```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("Library berhasil di-import!")
```

# Bagian 1 { Gambar Mobil }
# Membaca dan Menampilkan Gambar Asli

```python
# 202331214_ROMI FIRMANSYAH
img_mobil = cv2.imread('kendaraan.jpg')
    citra_mobil_gray = cv2.cvtColor(img_mobil, cv2.COLOR_BGR2GRAY)
        plt.imshow(citra_mobil_gray, cmap='gray')
        plt.title('Gambar Mobil Asli (Grayscale)')
        plt.show()
```

# Mean Filter

```python
# 202331214_ROMI FIRMANSYAH
kernel_mean = np.ones((5, 5), np.float32) / 25
citra_mobil_mean = cv2.filter2D(citra_mobil_gray, -1, kernel_mean)
    plt.imshow(citra_mobil_mean, cmap='gray')
    plt.title('Hasil Filter Rata-rata')
    plt.show()
```

# Median Filter

```python
# 202331214_ROMI FIRMANSYAH
citra_mobil_median = cv2.medianBlur(citra_mobil_gray, 5)
    plt.imshow(citra_mobil_median, cmap='gray')
    plt.title('Hasil Filter Median')
    plt.show()
```

# Edge Filter

```python
#202331214_ROMI FIRMANSYAH
sobel_x = cv2.Sobel(citra_mobil_gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(citra_mobil_gray, cv2.CV_64F, 0, 1, ksize=5)
    citra_mobil_tepi = np.sqrt(sobel_x**2 + sobel_y**2)
        plt.imshow(citra_mobil_tepi, cmap='gray')
        plt.title('Hasil Filter Batas (Deteksi Tepi)')
        plt.show()
```

# Bagian 2 {Gambar Boneka}
# Membaca & Menampilkan Gambar Boneka

```python
# 202331214_ROMI FIRMANSYAH
citra_boneka = cv2.imread('boneka.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(citra_boneka, cmap='gray')
plt.title('Gambar Boneka Asli (dengan Noise)')
plt.show()
```

# Konvolusi dengan Filter 2D untuk Menghilangkan Noise ya gaes ya!

```python
# 202331214_ROMI FIRMANSYAH
citra_boneka_bersih = cv2.medianBlur(citra_boneka, 5)
plt.imshow(citra_boneka_bersih, cmap='gray')
plt.title('Hasil Konvolusi Filter 2D (Median)')
plt.show()
```
